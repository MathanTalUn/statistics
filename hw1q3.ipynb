```python
import pandas as pd
path = "/home/appendicitis.csv"
df = pd.read_csv(path)
count_2 = (df['Pathology'] == 2).sum()
num_elements = len(df['Pathology'])
print(count_2/num_elements)
grouped = df.groupby(['Sex', 'Pathology']).size().unstack()
print(grouped)
count_sexFemale_pathology2 = ((df['Sex'] == 'Female') & (df['Pathology'] == 2)).sum()
num_sexFemale = len(df[df['Sex'] == 'Female'])
print(count_sexFemale_pathology2/num_sexFemale)
count_sexMale_pathology2 = ((df['Sex'] == 'Male') & (df['Pathology'] == 2)).sum()
num_sexMale = len(df[df['Sex'] == 'Male'])
print(count_sexMale_pathology2/num_sexMale)

import numpy as np
import matplotlib.pyplot as plt
male_ages = df[df['Sex'] == 'Male']['Age'].sort_values()
female_ages = df[df['Sex'] == 'Female']['Age'].sort_values()
male_ecdf = np.arange(1, len(male_ages) + 1) / len(male_ages)
female_ecdf = np.arange(1, len(female_ages) + 1) / len(female_ages)
plt.figure(figsize=(8, 6))
plt.plot(male_ages, male_ecdf, label='Male', linestyle='-', marker='', color='blue')
plt.plot(female_ages, female_ecdf, label='Female', linestyle='-', marker='', color='red')
plt.xlabel('age')
plt.ylabel('Cumulative Distribution')
plt.title('Empirical Cumulative Distribution of Age by Gender')
plt.legend()
plt.grid(True)
plt.show()
```
